**Standardize the dataset**

```python
scaler=MinMaxScaler()
```

```python
train_std=pd.DataFrame(scaler.fit_transform(train.iloc[:,0:25]))
```

```python
train_std.head()
```

** Applying PCA which covers 95% of the variance**

```python
pca=decomposition.PCA(0.95)
train_pca=pd.DataFrame(pca.fit_transform(train_std))
```

```python
pca.explained_variance_ratio_
```

Got 5 principal components covering 95% of the variance.   
First principal component covers 23.49% of the variance. 
Second principal component covers 19.58% of the variance.   
Third principal component covers 19.45% of the variance.   
Fourth principal component covers 18.19% of the variance.   
Fifth principal component covers 14.63% of the variance

```python
pca_df=pd.DataFrame(pca.components_)
```

```python
pca_df
```

This table tells us 'how much' the feature influences the principal components.  
So higher the value in absolute value, higher the influence on the principal component.

```python
pca_df.shape
```

```python
train_pca.head()
```

```python
final_train_pca=pd.concat([train_pca,train['winPlacePerc']],axis=1)
```

```python
final_train_pca.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from pandas.plotting import scatter_matrix
```

```python
X=final_train_pca.drop('winPlacePerc', axis=1)    #Dropping the winPlacePerc column
y=final_train_pca['winPlacePerc']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.4, random_state=0)
```

```python
splot=pd.plotting.scatter_matrix(X_train, figsize=(20,20))
```

```python
regressor= DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
df=pd.DataFrame({'Actual': y_test,'Predicted': y_pred})
df
```

```python
print("Mean Absolute error", metrics.mean_absolute_error(y_test, y_pred))
print("Mean Squared Error", metrics.mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
